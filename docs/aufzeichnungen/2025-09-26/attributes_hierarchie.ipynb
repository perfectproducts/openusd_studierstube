```python
from pxr import Usd, UsdGeom, UsdShade, Sdf, Gf
file_path = "_assets/attributes_homework.usda"
#
stage: Usd.Stage = Usd.Stage.CreateNew(file_path)
world_xform: UsdGeom.Xform = UsdGeom.Xform.Define(stage, "/World")
# Define a Scope prim in stage at `/Geometry`
geom_scope: UsdGeom.Scope = UsdGeom.Scope.Define(stage, "/World/Geometry")
# Define an Xform prim in the stage as a child of /Geometry called GroupTransform_Tree1
xform: UsdGeom.Xform = UsdGeom.Xform.Define(stage, geom_scope.GetPath().AppendPath("GroupTransform_Tree1"))
#
#Tree1
Treetop1: UsdGeom.Sphere = UsdGeom.Sphere.Define(stage, xform.GetPath().AppendPath("Treetop1"))
Treetop1_displaycolor: Usd.Attribute = Treetop1.GetDisplayColorAttr()
Treetop1_displaycolor.Set([(0.0, 1.0, 0.0)])
Treetop1_radius = Usd.Attribute = Treetop1.GetRadiusAttr()
Treetop1_radius.Set(1.5)
UsdGeom.XformCommonAPI(Treetop1).SetTranslate(Gf.Vec3d(2, 3, 1))
Treetrunk1: UsdGeom.Cylinder = UsdGeom.Cylinder.Define(stage, xform.GetPath().AppendPath("Treetrunk1"))
Treetrunk1_axis: Usd.Attribute = Treetrunk1.GetAxisAttr()
Treetrunk1_axis.Set("Y")
Treetrunk1_radius: Usd.Attribute = Treetrunk1.GetRadiusAttr()
Treetrunk1_radius.Set(0.2)
Treetrunk1_displaycolor: Usd.Attribute = Treetrunk1.GetDisplayColorAttr()
Treetrunk1_displaycolor.Set([(0.9, 0.1, 0.0)])
UsdGeom.XformCommonAPI(Treetrunk1).SetTranslate(Gf.Vec3d(2, 1, 1))
#
xform: UsdGeom.Xform = UsdGeom.Xform.Define(stage, geom_scope.GetPath().AppendPath("GroupTransform_Tree2"))
Treetop2: UsdGeom.Sphere = UsdGeom.Sphere.Define(stage, xform.GetPath().AppendPath("Treetop2"))
Treetop2_displaycolor: Usd.Attribute = Treetop2.GetDisplayColorAttr()
Treetop2_displaycolor.Set([(0.0, 1.0, 0.0)])
Treetop2_radius = Usd.Attribute = Treetop2.GetRadiusAttr()
Treetop2_radius.Set(1.2)
UsdGeom.XformCommonAPI(Treetop2).SetTranslate(Gf.Vec3d(-2, 3, 1.5))
Treetrunk2: UsdGeom.Cylinder = UsdGeom.Cylinder.Define(stage, xform.GetPath().AppendPath("Treetrunk2"))
Treetrunk2_axis: Usd.Attribute = Treetrunk2.GetAxisAttr()
Treetrunk2_axis.Set("Y")
Treetrunk2_radius: Usd.Attribute = Treetrunk2.GetRadiusAttr()
Treetrunk2_radius.Set(0.2)
Treetrunk2_displaycolor: Usd.Attribute = Treetrunk2.GetDisplayColorAttr()
Treetrunk2_displaycolor.Set([(0.9, 0.1, 0.0)])
UsdGeom.XformCommonAPI(Treetrunk2).SetTranslate(Gf.Vec3d(-2, 1, 1.5))
# 
xform: UsdGeom.Xform = UsdGeom.Xform.Define(stage, geom_scope.GetPath().AppendPath("GroupTransform_House"))
Building: UsdGeom.Cube = UsdGeom.Cube.Define(stage, xform.GetPath().AppendPath("House_Building"))
House_displaycolor: Usd.Attribute = Building.GetDisplayColorAttr()
House_displaycolor.Set([(0.0, 0.2, 0.7)])
UsdGeom.XformCommonAPI(Building).SetTranslate(Gf.Vec3d(0, 1, -1))
#
Roof: UsdGeom.Mesh = UsdGeom.Mesh.Define(stage, xform.GetPath().AppendPath("Roof"))
# Punkte (eine quadratische Basis + Spitze)
points = [
    Gf.Vec3f(-1, 0, -1),  # 0: Basis hinten links
    Gf.Vec3f( 1, 0, -1),  # 1: Basis hinten rechts
    Gf.Vec3f( 1, 0,  1),  # 2: Basis vorne rechts
    Gf.Vec3f(-1, 0,  1),  # 3: Basis vorne links
    Gf.Vec3f( 0, 2,  0)   # 4: Spitze nach +Y
]
# Flächen: Basis (1 Quadrat) + 4 Seitenflächen (Dreiecke)
faceVertexCounts = [4, 3, 3, 3, 3]
faceVertexIndices = [
    0, 1, 2, 3,  # Basis
    0, 1, 4,     # Seite 1
    1, 2, 4,     # Seite 2
    2, 3, 4,     # Seite 3
    3, 0, 4      # Seite 4
]
Roof.CreatePointsAttr(points)
Roof.CreateFaceVertexCountsAttr(faceVertexCounts)
Roof.CreateFaceVertexIndicesAttr(faceVertexIndices)
# Einfaches Material (RedStrong)
material = UsdShade.Material.Define(stage, Sdf.Path("/Looks/RedStrong"))
shader = UsdShade.Shader.Define(stage, Sdf.Path("/Looks/RedStrong/Shader"))
shader.CreateIdAttr("UsdPreviewSurface")
shader.CreateInput("diffuseColor", Sdf.ValueTypeNames.Color3f).Set((1.0, 0.0, 0.0))
shader.CreateInput("emissiveColor", Sdf.ValueTypeNames.Color3f).Set((0.2, 0.0, 0.0))
material.CreateSurfaceOutput().ConnectToSource(shader.ConnectableAPI(), "surface")
UsdShade.MaterialBindingAPI(Roof).Bind(material)
UsdGeom.XformCommonAPI(Roof).SetTranslate(Gf.Vec3d(0, 2, -1))
#
Floor: UsdGeom.Cube = UsdGeom.Cube.Define(stage, xform.GetPath().AppendPath("Floor"))
Floor.GetSizeAttr().Set(1.0) 
UsdGeom.XformCommonAPI(Floor).SetScale((8.0, 0.05, 5.0))  
UsdGeom.XformCommonAPI(Floor).SetTranslate(Gf.Vec3d(0, 0, 0))
Floor_size: Usd.Attribute = Floor.GetSizeAttr()
Floor_displaycolor: Usd.Attribute = Floor.GetDisplayColorAttr()
Floor_displaycolor.Set([(0.9, 0.3, 0.0)])
#
stage.Save()
```
